Dataset:- https://www.kaggle.com/datasets/shivam2503/diamonds

1. Bar --> Compare categorical data. Dont use it when showing trends over time. EX: Sales of different products

2. Line --> Show trends over time. Dont use it when displaying parts of a whole. EX: Stock prices over month

3. Pie Chart --> Displaying parts of a whole. Dont use it when representing distribution of continuous data. EX: Types of expenses in a budget

4. Histogram --> Representing distribution of continuous data. Dont use it when showing relationship between numerical vars. EX: Distibution of heights in a population

5. Scatter Plot --> Show relationships between numerical variables. Dont use it when displaying correlation in a matrix. EX: Sudy hours vs Exam Scores

6. Heatmap --> Display coreelation / concentration in a matrix. Dont use it when identifying outliers in data. EX: Customer preferences for products

7. Box Plot --> Identifying outliers and data distibution, EX: Salary distibution across job roles

8. Bubble Chart --> Disply three dimensions using bubbles. EX: Visualization GDP, Population and Life Expentancy

9. Radar Chart --> Show multivariate data on a 2-D Plane. EX: Skills of different job candidates

10. Polar Area Chart --> Display proportions of data categories. EX: Time spent on different activities

11. Tree Map --> Display hirarical data. EX: Visualizing market share of smartphone brands

12. Choropleth Map --> Representing data using geographical regions. EX: Population by State

13. Word Cloud --> Visualize text data by showing word frequency. Most frequent words in customer reviews

14. Funnel Chart --> Show sequential stages of a process. EX: Sales funnel analysis

15. Violin Plot --> Compare distribution of continuous variable across multiple categories. EX: Visualize the distribution of heights across different regions in India

```python
import pandas as pd
```

```python
data = pd.read_csv('diamonds.csv')

data
```

```python
data.info()
```

```python
import plotly.express as px
```

```python
fig = px.histogram(data, x='price')

fig.show()
```

```python
px.histogram(data, x='price',
             title='Histogram of Diamonds')
```

```python
px.histogram(data, x='price',
             title='Histogram of Diamonds',
             height=400, width=600)
```

```python
px.bar(data, x='cut')
```

```python
data['cut'].value_counts()
```

```python
cut = data['cut'].value_counts()

cut
```

```python
cut.index
```

```python
cut.values
```

```python
px.bar(cut,
       x=cut.index, y=cut.values)
```

```python
px.bar(cut, title='Bar Chart for Cut',
       x=cut.index, y=cut.values)
```

```python
px.bar(cut, title='Bar Chart for Cut',
       x=cut.index, y=cut.values,
       labels={'y':'Count',
               'cut':'Cut'})
```

```python
px.bar(cut, title='Bar Chart for Cut',
       x=cut.index, y=cut.values,
       labels={'y':'Count',
               'cut':'Cut'},
       color=cut.index)
```

```python
px.pie(names=cut.index,
       values=cut.values,
       title='Pie Chart')
```
